Standardize the predictor variables and redo the regression

```python
def lineplots(data, preds, response, subplots=True):
    pretty_nms = {'X3' : 'police funding','X4' : '25+ w/ 4 yrs of highschool',
                  'X5' : '16-19 not in highschool','X6' : '18-25 in college',
                  'X7' : '25+ w/ 4 yrs in collegue', 'X2' : '# crimes', 
                  'Ystd' : '# crimes (stdized)'};

    npts, _ = data.shape
    data_plot = data.rename(columns=pretty_nms)
    nm1 = pretty_nms[response]
    nms = [pretty_nms[nm] for nm in preds]
    if subplots:
        fig, axes = plt.subplots(figsize=(6,8), nrows=2, ncols=1)
        data_plot[nm1].plot.line(ax=axes[0])
        data_plot[nms].plot.line(ax=axes[1])
    else:
        data_plot[nms + [nm1]].plot.line()
```

```python
lineplots(data, preds, response)
```

```python
data2 = data[['X1', 'X2']].copy()
for nm in preds:
    data2[nm] = (data[nm] - data[nm].mean()) / data[nm].std()
data2['Ystd'] = (data[response] - data[response].mean()) / data[response].std()
data2.head()    
```
